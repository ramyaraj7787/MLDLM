### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
%matplotlib inline
```

```python
irisData = datasets.load_iris()
```

```python
x = irisData.data[:, :3] 
y = irisData.target
```

```python
pca = decomposition.PCA()
dec = pca.fit_transform(x)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter3D(dec[:,0], dec[:,1], dec[:,2], c=y)
ax.set_title("3D Plot")
```
